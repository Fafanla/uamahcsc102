```python
#project 1 
from abc import ABC 


class coup_de_escariva (ABC):
    
    def Fan_page(self):
        pass
    
class FC_Cirok(coup_de_escariva):
    def Fan_page(self):
        print("Welcome to FC Cirok")
class Madiba_FC(coup_de_escariva):
    def Fan_page(self):
        print("Welcome to Madiba FC")
class Blue_Jay_FC(coup_de_escariva):
    def Fan_page(self):
        print("Welcome to Blue Jay FC")
class TSG_Walker(coup_de_escariva):
    def Fan_page(self):
        print("Welcome to TSG Walker FC")
        

club = input("Enter your club's name  ")
if club == "Cirok":
    fan_page = FC_Cirok()
elif club == "Madiba":
    fan_page = Madiba_FC()
elif club == "Blue Jay":
    fan_page = Blue_Jay_FC()
elif club == "TSG Walker":
    fan_page = TSG_Walker()
else:
    print("Invalid club selected")
        
        
        
fan_page.Fan_page()       
        
        
        
        
        
    
    
```
